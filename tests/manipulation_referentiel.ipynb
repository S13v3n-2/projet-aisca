```python
# === Step 1: Import libraries ===
from sentence_transformers import SentenceTransformer, util
import numpy as np
import pandas as pd
import json

with open("../data/referentiel.json", 'r', encoding='utf-8') as f:
    data = json.load(f)

# 2. Création de l'index des compétences
# On itère sur data['blocs'] qui est la liste des blocs
competences_index = {
    comp['id']: comp
    for bloc in data['blocs']
    for comp in bloc['competences']
}

# Test d'accès
print(f"Nombre de compétences indexées : {len(competences_index)}")
```

```python
from sentence_transformers import SentenceTransformer
import json
import numpy as np

# 1. Charger le modèle SBERT (all-MiniLM-L6-v2 est recommandé dans ton sujet)
# C'est un modèle léger, rapide et performant en local [cite: 323]
model = SentenceTransformer('all-MiniLM-L6-v2')
# 4. Générer les embeddings
```

```python
textes = [c['texte'] for c in competences_index.values()]
embeddings = model.encode(textes, convert_to_tensor=True)

# 5. Stocker les résultats pour le moteur de recherche
# On associe chaque ID à son vecteur
vector_store = {
    comp['id']: embeddings[i]
    for i, comp in enumerate(competences_index.values())
}

print(f"Référentiel vectorisé : {len(vector_store)} compétences prêtes.")
```

```python
from sentence_transformers import util

def calculer_similarites(cheminentree_utilisateur, vector_store, model):
    """
    Compare une saisie utilisateur avec toutes les compétences indexées.
    """
    # 1. Transformer la saisie utilisateur en vecteur (Embedding)
    user_embedding = model.encode(entree_utilisateur, convert_to_tensor=True)

    scores = []

    # 2. Comparer avec chaque compétence du référentiel
    for comp_id, comp_embedding in vector_store.items():
        # Calcul de la similarité cosinus (résultat entre 0 et 1)
        score = util.cos_sim(user_embedding, comp_embedding).item()

        scores.append({
            "id": comp_id,
            "score": round(score, 4)
        })

    # 3. Trier par score décroissant pour avoir les meilleures correspondances
    scores = sorted(scores, key=lambda x: x['score'], reverse=True)
    return scores

# --- TEST ---
input_test = "Je sais piloter des campagnes publicitaires sur Facebook Ads et optimiser le référencement naturel des sites web."
resultats = calculer_similarites(input_test, vector_store, model)

print(f"Top 20 correspondances pour : '{input_test}'")
text_cross = [competences_index[res['id']]['texte'] for res in resultats[:20]]
for res in resultats[:20]:
    # On utilise l'index de compétences créé précédemment pour afficher le texte
    texte = competences_index[res['id']]['texte']
    print(f"-{competences_index[res['id']]['id']} {texte} (Score: {res['score']})")
```

```python
from sentence_transformers import CrossEncoder

# 1. Load a pretrained CrossEncoder model
model_cross = CrossEncoder("cross-encoder/ms-marco-MiniLM-L6-v2")
# The texts for which to predict similarity scores
query = input_test
passages = text_cross

# 2a. Either predict scores pairs of texts
scores = model_cross.predict([(query, passage) for passage in passages])
#print(scores)
# => [8.607139 5.506266 6.352977]

# 2b. Or rank a list of passages for a query
ranks = model_cross.rank(query, passages, return_documents=True)

import math

def sigmoid(x):
    return 1 / (1 + math.exp(-x))
# Exemple pour ton top score (2.22)

print("Query:", query)
for rank in ranks[:3]:
    print(f"- #{rank['corpus_id']} ({rank['score']:.2f}): {rank['text']}")
```
